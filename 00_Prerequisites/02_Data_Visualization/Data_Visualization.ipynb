```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Line Plots

```python
fifa_data = pd.read_csv('fifa.csv')
fifa_data.head()
```

```python
plt.figure(figsize=(15,5))

plt.title('Fifa Data')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['ARG'], label = 'ARG')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['BRA'], label = 'BRA')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['ESP'], label = 'ESP')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['FRA'], label = 'FRA')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['GER'], label = 'GER')
sns.lineplot(data=fifa_data,x=fifa_data['Date'],y = fifa_data['ITA'], label = 'ITA')

plt.xlabel('Date')
plt.ylabel('Country')
```

## Bar Charts and HeatMaps

```python
flights = pd.read_csv('flight_delays.csv')
flights.head()
```

```python
plt.figure(figsize = (15,5))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x = flights['Month'], y = flights['VX'])

plt.xlabel('Month')
plt.ylabel('Delay')
```

```python
plt.figure(figsize=(15,5))
plt.title('Average Arrival Delay for Each Airline, by Month')

sns.heatmap(data=flights)

plt.xlabel('Country')
plt.ylabel('Month')
```

## Scatter Plot

```python
insurance = pd.read_csv('insurance.csv')
insurance.head()
```

```python
sns.scatterplot(x = insurance['bmi'],y=insurance['charges'])
```

```python
sns.regplot(x = insurance['bmi'],y=insurance['charges'])
```

```python
sns.scatterplot(x = insurance['bmi'],y=insurance['charges'],hue =insurance['region'])
```

```python
sns.lmplot(x = 'bmi',y='charges',hue ='region',data=insurance)
```

```python
sns.swarmplot(x=insurance['smoker'],y=insurance['charges'])
```

## Distributions

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.distplot(a = iris['Petal Length (cm)'])
```

```python
sns.kdeplot(data=iris['Petal Width (cm)'])
```

```python
sns.jointplot(x=iris['Petal Length (cm)'], y=iris['Sepal Width (cm)'], kind="kde")
```
